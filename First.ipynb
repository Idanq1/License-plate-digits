```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(r"..\Images\original_lane_detection_5.jpg")
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# scale_pcnt = 60
# width = int(img.shape[1] * scale_pcnt / 100)
# height = int(img.shape[0] * scale_pcnt / 100)
# dim = (width, height)
# img = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)

# ROI
height, width = img.shape[:2]
factor = 0.6
start_y = int(height * (1 - factor))
img_roi = img[start_y:height, 0:width]


hsv = cv2.cvtColor(img_roi, cv2.COLOR_BGR2HSV)
lower_white = np.array([0, 0, 203])
upper_white = np.array([180, 43, 255])
white_mask = cv2.inRange(hsv, lower_white, upper_white)

lower_yellow = np.array([179, 254, 254])
upper_yellow = np.array([180, 255, 255])
yellow_mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

# Combine the two masks
combined_mask = cv2.bitwise_or(white_mask, yellow_mask)
hsv_mask = cv2.bitwise_and(img_roi, img_roi, mask=combined_mask)

kernel_size = 2  # Example size, may need adjustment
kernel = np.ones((kernel_size, kernel_size), np.uint8)
opening = cv2.morphologyEx(hsv_mask, cv2.MORPH_OPEN, kernel)

# Blend with original image
# blend_factor = 0.5  # Adjust this factor between 0 (only original) and 1 (only morph_image)
# opening = cv2.addWeighted(hsv_mask, 1 - blend_factor, opening, blend_factor, 0)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(100, 50))  # 1 row, 2 columns
ax1.imshow(cv2.cvtColor(opening, cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(hsv_mask, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_g = cv2.cvtColor(opening, cv2.COLOR_BGR2GRAY)
img_b = cv2.GaussianBlur(img_g, (5, 5), 0)

plt.imshow(img_b, cmap="gray")
```

```python
low_thresh = 50
high_thresh = 150
edges = cv2.Canny(img_b, low_thresh, high_thresh)
plt.imshow(edges, cmap="gray")
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=50, minLineLength=50, maxLineGap=50)

line_image = np.zeros_like(img_roi, dtype=np.uint8)

if lines is not None:
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(line_image, (x1, y1), (x2, y2), (0, 0, 255), 1)

combo_image = cv2.addWeighted(img_roi, 0.8, line_image, 1, 2)
plt.imshow(cv2.cvtColor(combo_image, cv2.COLOR_BGR2RGB))
len(lines)
```
